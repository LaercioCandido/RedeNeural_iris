```python
import numpy as np
from sklearn import datasets
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import StandardScaler

np.random.seed(5)

iris = datasets.load_iris()
X = iris.data
y = iris.target

#normalizaçao
scaler = StandardScaler()
scaler.fit(X)
X = scaler.transform(X)

sufInd = np.arange(150) #gera uma sequencia de numeros aleatorios entre 0 e 149 (pq o dataset iris tem 150 instancias)
np.random.shuffle(sufInd)

#split em treino e teste
X_train = X[sufInd[:100], :]
X_test = X[sufInd[100:], :]
y_train = y[sufInd[:100]]
y_test = y[sufInd[100:]]

clf = MLPClassifier(solver='sgd', hidden_layer_sizes=(3,), learning_rate_init=0.7, activation='logistic', max_iter=1500, random_state=1) #stochastic gradient descent
clf.fit(X_train, y_train)
y_pred = clf.predict_proba(X_test)
y_pred_aux = np.argmax(y_pred, 1) #procura em cada linha qual é a coluna q possui a maior probabilidade
accuracy_score(y_test, y_pred_aux)
```
